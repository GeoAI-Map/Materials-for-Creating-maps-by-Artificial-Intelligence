```python
import geopandas as gpd
import folium

# Load your shapefiles
boundary_path = './boundary.shp'  # Update the path to your boundary shapefile
mainlandburn_path = './mainlandburn.shp'  # Update the path to your mainlandburn shapefile

boundary = gpd.read_file(boundary_path)
mainlandburn = gpd.read_file(mainlandburn_path)

# Merge attribute data and geometries as needed
# Ensure CRS matches and correct handling of geometries if required
mainlandburn['geometry'] = mainlandburn.set_crs(boundary.crs, allow_override=True).geometry

# Calculate centroid for map initialization
centroid = mainlandburn.geometry.centroid.to_crs(epsg=4326).mean()

# Initialize Folium map at the calculated centroid
m = folium.Map(location=[centroid.y, centroid.x], zoom_start=7, tiles='cartodbpositron')

# Create a choropleth map
folium.Choropleth(
    geo_data=mainlandburn,
    data=mainlandburn,
    columns=['NAME_1', 'Burned_Are'],
    key_on='feature.properties.NAME_1',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Average Burned Area (2002-2022)'
).add_to(m)

# Overlay state boundaries
folium.GeoJson(
    boundary,
    name='State Boundaries',
    style_function=lambda x: {'color': 'black', 'weight': 1.5, 'fillOpacity': 0}
).add_to(m)

# Add tooltips
folium.GeoJson(
    mainlandburn,
    name='Tooltip Info',
    style_function=lambda x: {'fillColor': 'transparent', 'color': 'transparent'},
    tooltip=folium.GeoJsonTooltip(
        fields=['NAME_1', 'Burned_Are'],
        aliases=['Region:', 'Fire Count:'],
        localize=True
    )
).add_to(m)

# Add layer control
folium.LayerControl().add_to(m)

# Save or show the map
m
```
