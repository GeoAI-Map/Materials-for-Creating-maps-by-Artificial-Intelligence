```python
import pandas as pd
import plotly.express as px
import folium
from plotly.io import to_html
import geopandas as gpd
import folium
from branca.colormap import linear

# Paths to the shapefiles
boundary_path = './boundary.shp'
mainlandburn_path = './mainlandburn.shp'
# Load the shapefiles
boundary_gdf = gpd.read_file(boundary_path)
mainlandburn_gdf = gpd.read_file(mainlandburn_path)

# Ensure CRS is set
if not boundary_gdf.crs:
    boundary_gdf.set_crs("EPSG:4326", inplace=True)
if not mainlandburn_gdf.crs:
    mainlandburn_gdf.set_crs("EPSG:4326", inplace=True)

# Create a map centered on Portugal
map_center = [39.557191, -7.8536599]  # Approximate center of Portugal
folium_map = folium.Map(location=map_center, zoom_start=7, tiles='cartodbpositron', control_scale=True)

# Define a color scale for the choropleth based on the Burned_Are values using a warm color scheme
color_scale = linear.YlOrRd_09.scale(mainlandburn_gdf['Burned_Are'].min(), mainlandburn_gdf['Burned_Are'].max())

# Calculate interval classes for the choropleth
interval = (mainlandburn_gdf['Burned_Are'].max() - mainlandburn_gdf['Burned_Are'].min()) / 5
classes = [mainlandburn_gdf['Burned_Are'].min() + i * interval for i in range(6)]

# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 10px; width: 300px; background-color: rgba(255,255,255,0.8); padding: 10px; border-radius: 8px;">
                 <h3 style="font-size:20px; font-family:Serif; font-weight: bold;">AVERAGE BURNED AREA PER REGION AREA</h3>
                 <h4 style="font-size:16px; font-family:Serif;">Mainland of Portugal, 2002-2022</h4>
             </div>
             <div style="position: absolute; bottom: 20px; left: 10px; width: 300px; color: grey; font-size:10px; font-family:Arial;">
                 <p>Author: Parinda Pannoon<br>Date: 13/04/2024<br>Spatial Reference: EPSG 3857<br>Data source: Global Wildfire Information System (GWIS)</p>
             </div>
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))
# Add choropleth layer for the burned areas
folium.Choropleth(
    geo_data=mainlandburn_gdf.to_json(),
    name='Choropleth',
    data=mainlandburn_gdf,
    columns=['NAME_1', 'Burned_Are'],
    key_on='feature.properties.NAME_1',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Average Burned Area per Region Area',
    bins=classes,
    reset=True,
    highlight=True
).add_to(folium_map)

# Add tooltips for NAME_1 names and burned areas
folium.features.GeoJson(
    mainlandburn_gdf,
    name='Labels',
    style_function=lambda x: {'color':'transparent', 'fillColor':'transparent', 'weight':0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1', 'Burned_Are'],
        aliases=['Region:', 'Average Burned Area:'],
        localize=True
    )
).add_to(folium_map)

# Adding labels to each region
for _, region in mainlandburn_gdf.iterrows():
    # Ensure the point exists within the geometry
    point = region['geometry'].centroid
    folium.Marker(
        [point.y, point.x],
        icon=folium.DivIcon(
            html=f'<div style="font-size:12pt; color:black; -webkit-text-stroke: 0.5px white;">{region["NAME_1"]}</div>'
        )
    ).add_to(folium_map)

# Add state boundaries layer with brown color
folium.features.GeoJson(
    boundary_gdf,
    name='State Boundaries',
    style_function=lambda x: {'color': 'brown', 'weight': 1, 'fillOpacity': 0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1'],
        aliases=['State:'],
        localize=True
    )
).add_to(folium_map)



# Add layer control
folium.LayerControl().add_to(folium_map)



# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 50px; width: 300px; z-index: 1000; font-family: Serif;">
                 <h3 style="font-size:24px; font-weight: bold; margin: 0;">AVERAGE BURNED AREA PER REGION</h3>
                 <h4 style="font-size:18px; margin: 0;">Mainland of Portugal, 2002-2022</h4>
             </div>
             <div style="position: absolute; bottom: 10px; right: 50px; z-index: 1000; font-size:10px; font-family:Arial; color: grey;">
                 <p>Author: Parinda Pannoon<br>Date: 13/04/2024<br>Spatial Reference: EPSG 3857<br>Data source: Global Wildfire Information System (GWIS)</p>
             </div>
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))

# Load your CSV data
data = pd.read_csv('D:\CDE_SEMESTER3\Thesis\data\data_choropleth\MCD64.006.yearly-ba-nf.2002-2022.PRT_Portugal.csv')

# Assuming the data has columns 'Year' and 'Burned_Area'
fig = px.bar(data, x='Year', y='Burned Area [ha]', 
             title='Average Burned Area Yearly',
             labels={'Burned Area [ha]': 'Burned Area (ha)'},
             text='Burned Area [ha]')

# Customizing the chart appearance
fig.update_traces(marker_color='#E1C461', textposition='outside')
fig.update_layout(title_font=dict(size=18, family='Serif', color='black'),
                  title_x=0.5,
                  xaxis_title='Year',
                  yaxis_title='Average Burned Area (ha)',
             
                  plot_bgcolor='rgba(0,0,0,0)',
         
                  height=250)

# Convert Plotly figure to an HTML string for embedding
plotly_html = to_html(fig, full_html=False, include_plotlyjs='cdn')


# Embed the Plotly chart using custom HTML and CSS
folium_map.get_root().html.add_child(folium.Element(f"""
<div style="position: fixed; top: 10px; right: 10px; width: 300px; height: 250px; z-index: 9999; overflow: hidden;">
    {plotly_html}
</div>
"""))

# Save and display the map
folium_map.save('Interactive_Map_with_Embedded_Chart.html')
folium_map
```

```python
import pandas as pd
import plotly.express as px
import folium
from plotly.io import to_html
import geopandas as gpd
import folium
from branca.colormap import linear

# Paths to the shapefiles
boundary_path = './boundary.shp'
mainlandburn_path = './mainlandburn.shp'
# Load the shapefiles
boundary_gdf = gpd.read_file(boundary_path)
mainlandburn_gdf = gpd.read_file(mainlandburn_path)

# Ensure CRS is set
if not boundary_gdf.crs:
    boundary_gdf.set_crs("EPSG:4326", inplace=True)
if not mainlandburn_gdf.crs:
    mainlandburn_gdf.set_crs("EPSG:4326", inplace=True)

# Create a map centered on Portugal
map_center = [39.557191, -7.8536599]  # Approximate center of Portugal
folium_map = folium.Map(location=map_center, zoom_start=7, tiles='cartodbpositron', control_scale=True)

# Define a color scale for the choropleth based on the Burned_Are values using a warm color scheme
color_scale = linear.YlOrRd_09.scale(mainlandburn_gdf['Burned_Are'].min(), mainlandburn_gdf['Burned_Are'].max())

# Calculate interval classes for the choropleth
interval = (mainlandburn_gdf['Burned_Are'].max() - mainlandburn_gdf['Burned_Are'].min()) / 5
classes = [mainlandburn_gdf['Burned_Are'].min() + i * interval for i in range(6)]

# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 10px; width: 300px; background-color: rgba(255,255,255,0.8); padding: 10px; border-radius: 8px;">
                 <h3 style="font-size:20px; font-family:Serif; font-weight: bold;">AVERAGE BURNED AREA PER REGION AREA</h3>
                 <h4 style="font-size:16px; font-family:Serif;">Mainland of Portugal, 2002-2022</h4>
             </div>
             <div style="position: absolute; bottom: 20px; left: 10px; width: 300px; color: grey; font-size:10px; font-family:Arial;">
                 <p>Author: Parinda Pannoon<br>Date: 13/04/2024<br>Spatial Reference: EPSG 3857<br>Data source: Global Wildfire Information System (GWIS)</p>
             </div>
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))
# Add choropleth layer for the burned areas
folium.Choropleth(
    geo_data=mainlandburn_gdf.to_json(),
    name='Choropleth',
    data=mainlandburn_gdf,
    columns=['NAME_1', 'Burned_Are'],
    key_on='feature.properties.NAME_1',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Average Burned Area per Region Area',
    bins=classes,
    reset=True,
    highlight=True
).add_to(folium_map)

# Add tooltips for NAME_1 names and burned areas
folium.features.GeoJson(
    mainlandburn_gdf,
    name='Labels',
    style_function=lambda x: {'color':'transparent', 'fillColor':'transparent', 'weight':0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1', 'Burned_Are'],
        aliases=['Region:', 'Average Burned Area:'],
        localize=True
    )
).add_to(folium_map)

# Adding labels to each region
for _, region in mainlandburn_gdf.iterrows():
    # Ensure the point exists within the geometry
    point = region['geometry'].centroid
    folium.Marker(
        [point.y, point.x],
        icon=folium.DivIcon(
            html=f'<div style="font-size:8pt; color:black; -webkit-text-stroke: 0.5px white;">{region["NAME_1"]}</div>'
        )
    ).add_to(folium_map)

# Add state boundaries layer with brown color
folium.features.GeoJson(
    boundary_gdf,
    name='State Boundaries',
    style_function=lambda x: {'color': 'brown', 'weight': 1, 'fillOpacity': 0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1'],
        aliases=['State:'],
        localize=True
    )
).add_to(folium_map)



# Add layer control
folium.LayerControl().add_to(folium_map)



# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 50px; width: 300px; z-index: 1000; font-family: Serif;">
                 <h3 style="font-size:24px; font-weight: bold; margin: 0;">AVERAGE BURNED AREA PER REGION</h3>
                 <h4 style="font-size:18px; margin: 0;">Mainland of Portugal, 2002-2022</h4>
             </div>
         
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))
# HTML to incorporate the external link and style the credit text
credits_html = """
<div style="position: fixed; bottom: 0px; right: 0px; width: 300px; z-index: 9999; text-align: right; font-size: 10px;">
    Data source: <a href='https://gwis.jrc.ec.europa.eu/apps/country.profile/overview' target='_blank'>Global Wildfire Information System (GWIS)</a><br>
    Author: Your Name<br>
    Date: 2024<br>
    Spatial Reference: EPSG 3857
</div>
"""

# Embed the HTML using a Folium Element
folium_map.get_root().html.add_child(folium.Element(credits_html))
# Load your CSV data
data = pd.read_csv('D:\CDE_SEMESTER3\Thesis\data\data_choropleth\MCD64.006.yearly-ba-nf.2002-2022.PRT_Portugal.csv')

# Assuming the data has columns 'Year' and 'Burned_Area'
fig = px.bar(data, x='Year', y='Burned Area [ha]', 
             title='Average Burned Area Yearly',
             labels={'Burned Area [ha]': 'Burned Area (ha)'},
             text='Burned Area [ha]')

# Customizing the chart appearance
fig.update_traces(marker_color='#E1C461', textposition='outside')
fig.update_layout(title_font=dict(size=18, family='Serif', color='black'),
                  title_x=0.5,
                  xaxis_title='Year',
                  yaxis_title='Average Burned Area (ha)',
             
                  plot_bgcolor='rgba(0,0,0,0)',
         
                  height=250)

# Convert Plotly figure to an HTML string for embedding
plotly_html = to_html(fig, full_html=False, include_plotlyjs='cdn')


# Embed the Plotly chart using custom HTML and CSS
folium_map.get_root().html.add_child(folium.Element(f"""
<div style="position: fixed; top: 180px; right: 10px; width: 300px; height: 250px; z-index: 9999; overflow: hidden;">
    {plotly_html}
</div>
"""))

# Save and display the map
folium_map.save('Interactive_Map_with_Embedded_Chart.html')
folium_map
```

```python
import pandas as pd
import plotly.express as px
import folium
from plotly.io import to_html
import geopandas as gpd
import folium
from branca.colormap import linear

# Paths to the shapefiles
boundary_path = './boundary.shp'
mainlandburn_path = './mainlandburn.shp'
# Load the shapefiles
boundary_gdf = gpd.read_file(boundary_path)
mainlandburn_gdf = gpd.read_file(mainlandburn_path)

# Ensure CRS is set
if not boundary_gdf.crs:
    boundary_gdf.set_crs("EPSG:4326", inplace=True)
if not mainlandburn_gdf.crs:
    mainlandburn_gdf.set_crs("EPSG:4326", inplace=True)

# Create a map centered on Portugal
map_center = [39.557191, -7.8536599]  # Approximate center of Portugal
folium_map = folium.Map(location=map_center, zoom_start=7, tiles='cartodbpositron', control_scale=True)

# Define a color scale for the choropleth based on the Burned_Are values using a warm color scheme
color_scale = linear.YlOrRd_09.scale(mainlandburn_gdf['Burned_Are'].min(), mainlandburn_gdf['Burned_Are'].max())

# Calculate interval classes for the choropleth
interval = (mainlandburn_gdf['Burned_Are'].max() - mainlandburn_gdf['Burned_Are'].min()) / 5
classes = [mainlandburn_gdf['Burned_Are'].min() + i * interval for i in range(6)]

# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 10px; width: 300px; background-color: rgba(255,255,255,0.8); padding: 10px; border-radius: 8px;">
                 <h3 style="font-size:20px; font-family:Serif; font-weight: bold;">AVERAGE BURNED AREA PER REGION AREA</h3>
                 <h4 style="font-size:16px; font-family:Serif;">Mainland of Portugal, 2002-2022</h4>
             </div>
             <div style="position: absolute; bottom: 20px; left: 10px; width: 300px; color: grey; font-size:10px; font-family:Arial;">
                 <p>Author: Parinda Pannoon<br>Date: 13/04/2024<br>Spatial Reference: EPSG 3857<br>Data source: Global Wildfire Information System (GWIS)</p>
             </div>
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))
# Add choropleth layer for the burned areas
folium.Choropleth(
    geo_data=mainlandburn_gdf.to_json(),
    name='Choropleth',
    data=mainlandburn_gdf,
    columns=['NAME_1', 'Burned_Are'],
    key_on='feature.properties.NAME_1',
    fill_color='YlOrRd',
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name='Average Burned Area per Region Area',
    bins=classes,
    reset=True,
    highlight=True
).add_to(folium_map)

# Add tooltips for NAME_1 names and burned areas
folium.features.GeoJson(
    mainlandburn_gdf,
    name='Labels',
    style_function=lambda x: {'color':'transparent', 'fillColor':'transparent', 'weight':0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1', 'Burned_Are'],
        aliases=['Region:', 'Average Burned Area:'],
        localize=True
    )
).add_to(folium_map)

# Adding labels to each region
for _, region in mainlandburn_gdf.iterrows():
    # Ensure the point exists within the geometry
    point = region['geometry'].centroid
    folium.Marker(
        [point.y, point.x],
        icon=folium.DivIcon(
             html=f'<div style="font-size:8pt; color:black; -webkit-text-stroke: 1.5px white;">{region["NAME_1"]}</div>'
        )
    ).add_to(folium_map)

# Add state boundaries layer with brown color
folium.features.GeoJson(
    boundary_gdf,
    name='State Boundaries',
    style_function=lambda x: {'color': 'brown', 'weight': 1, 'fillOpacity': 0},
    tooltip=folium.features.GeoJsonTooltip(
        fields=['NAME_1'],
        aliases=['State:'],
        localize=True
    )
).add_to(folium_map)



# Add layer control
folium.LayerControl().add_to(folium_map)



# Adding titles and credits
title_html = '''
             <div style="position: absolute; top: 10px; left: 50px; width: 300px; z-index: 1000; font-family: Serif;">
                 <h3 style="font-size:24px; font-weight: bold; margin: 0;">AVERAGE BURNED AREA PER REGION</h3>
                 <h4 style="font-size:18px; margin: 0;">Mainland of Portugal, 2002-2022</h4>
             </div>
         
             '''
folium_map.get_root().html.add_child(folium.Element(title_html))
# HTML to incorporate the full credit information and adjust the position
credits_html = """
<div style="position: fixed; bottom: 50px; right: 10px; width: 300px; z-index: 9999; text-align: right; font-size: 10px;">
    Author: Parinda Pannoon<br>
    Date: 12/04/2024<br>
    Spatial Reference: EPSG 3857<br>
    Data source: <a href='https://gwis.jrc.ec.europa.eu/apps/country.profile/overview' target='_blank'>Global Wildfire Information System (GWIS)</a>
</div>
"""

# Embed the HTML using a Folium Element
folium_map.get_root().html.add_child(folium.Element(credits_html))
# Load your CSV data
data = pd.read_csv('D:\CDE_SEMESTER3\Thesis\data\data_choropleth\MCD64.006.yearly-ba-nf.2002-2022.PRT_Portugal.csv')

# Assuming the data has columns 'Year' and 'Burned_Area'
fig = px.bar(data, x='Year', y='Burned Area [ha]', 
             title='Average Burned Area Yearly',
             labels={'Burned Area [ha]': 'Burned Area (ha)'},
             text='Burned Area [ha]')

# Customizing the chart appearance
fig.update_traces(marker_color='#E1C461', textposition='outside')
fig.update_layout(title_font=dict(size=18, family='Serif', color='black'),
                  title_x=0.5,
                  xaxis_title='Year',
                  yaxis_title='Average Burned Area (ha)',
                 plot_bgcolor='rgba(0,0,0,0)',  # Transparent background
                 paper_bgcolor='rgba(0,0,0,0)',
         
                  height=250)

# Convert Plotly figure to an HTML string for embedding
plotly_html = to_html(fig, full_html=False, include_plotlyjs='cdn')


# Embed the Plotly chart using custom HTML and CSS
folium_map.get_root().html.add_child(folium.Element(f"""
<div style="position: fixed; top: 180px; right: 10px; width: 300px; height: 250px; z-index: 9999; overflow: hidden;">
    {plotly_html}
</div>
"""))
map_save_path = 'D:\CDE_SEMESTER3\Thesis\data\data_choropleth\Choro4thad.html'  # Adjust this to your preferred save location
folium_map.save(map_save_path)

folium_map
```
